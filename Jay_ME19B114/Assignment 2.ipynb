# Task 1: Edge Detection using Difference of Gaussians

```python
import cv2 
import matplotlib.pyplot as plt
```

```python
ein= cv2.imread("einstein.jpg")
plt.figure()
plt.imshow(ein,cmap='gray'),plt.axis("off"),plt.title("Original Image");
```

```python
img= cv2.imread("einstein.jpg")
gausblur5= cv2.GaussianBlur(img,(5,5),0)
gausblur9= cv2.GaussianBlur(img,(9,9),0)

plt.figure()
plt.imshow(img),plt.title("original"),plt.axis("off");
plt.figure()
plt.imshow(gausblur5),plt.axis("off"),plt.title("5*5 blur");
plt.figure()
plt.imshow(gausblur9),plt.axis("off"),plt.title("9*9 blur");
```

```python
diff = gausblur9-gausblur5
plt.figure()
plt.imshow(diff),plt.axis("off");
```

```python
barr = cv2.imread("barr.JPG")
plt.subplot(121),plt.imshow(cv2.cvtColor(barr,cv2.COLOR_BGR2RGB)),plt.axis('off'),plt.title('Original');
barr = cv2.cvtColor(barr, cv2.COLOR_BGR2HSV)
plt.subplot(122),plt.imshow(barr),plt.axis('off'),plt.title('hsv');
#taking an image of barrel from net
```

```python
his= cv2.calcHist([barr],[0],None,[180],[0,180]);

nor= cv2.normalize(his, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

_,nor = cv2.threshold(nor, 30, 255, cv2.THRESH_BINARY)
res = cv2.calcBackProject([frame], [0], nor, [0, 180], 1) # Back Projection
plt.imshow(res), plt.axis("off"), plt.title("res");
```
